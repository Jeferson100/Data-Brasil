```python
from bcb import Expectativas
def dados_expectativas_focus(
    indicador="IPCA",
    tipo_expectativa="ExpectativaMercadoMensais",
    data_inicio="2000-01-01",
    
):
    # End point
    em = Expectativas()
    ep = em.get_endpoint(tipo_expectativa)
    # Dados do IPCA
    atributos_necessarios = ["Indicador", "Data", "Media", "Mediana", "DataReferencia", "baseCalculo"]
    if all(hasattr(ep, attr) for attr in atributos_necessarios):
        ipca_expec = (
            ep.query()
            .filter(ep.Indicador == indicador)
            .filter(ep.Data >= data_inicio)
            .filter(ep.baseCalculo == 0)
            .select(
                ep.Indicador,
                ep.Data,
                ep.Media,
                ep.Mediana,
                ep.DataReferencia,
                ep.baseCalculo,
            )
            .collect()
        )
        return ipca_expec
    else:
        print("Indicador não encontrado")
```

```python
def dados_expectativas_focus(
    indicador="IPCA",
    tipo_expectativa="ExpectativaMercadoMensais",
    data_inicio="2000-01-01",
):
    # End point
    em = Expectativas()
    ep = em.get_endpoint(tipo_expectativa)

    # Dados do IPCA
    ipca_expec = (
        ep.query()
        .filter(getattr(ep, 'Indicador', None) == indicador)
        .filter(getattr(ep, 'Data', None) >= data_inicio)
        .filter(getattr(ep, 'baseCalculo', None) == 0)
        .select(
            getattr(ep, 'Indicador', None),
            getattr(ep, 'Data', None),
            getattr(ep, 'Media', None),
            getattr(ep, 'Mediana', None),
            getattr(ep, 'DataReferencia', None),
            getattr(ep, 'baseCalculo', None),
        )
        .collect()
    )
    return ipca_expec
```
